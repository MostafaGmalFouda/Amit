18. Sort a tuple by its float element

```python
items = (("item1", 10.5), ("item2", 5.7), ("item3", 8.9))

# TODO: sort by the float value
# your code here
# method sorted() return a new list containing all items from the iterable in ascending order
# we use keym method to sort by float value
sorted_items = sorted(items,key=lambda items:items[1])
print(sorted_items)
# Hint: Use sorted() with a key method.
```
